## Fehler beheben mit Pandas

Es gibt Funktionen die fehlerhafte Werte erkennen können. Danach können wir all diese Werte ersetzen durch valide Werte.

Viele dieser Funktionen kommen von Numpy

```python
import numpy as np
import pandas as pd
```

```python
pd.isnull(np.NaN)  # Alternative: isna() -> selbe Funktion
```

```python
pd.notnull(np.NaN)  # Alternative notna() -> selbe Funktion
```

Diese Funktionen funktionieren auch mit DataFrames (und Serien)

```python
serie = pd.Series([1, np.NaN, 3])
```

```python
serie
```

```python
pd.isnull(serie)
```

```python
pd.notnull(serie)
```

```python
serie[pd.isnull(serie)] = 0  # Bei NaN Werten 0 eintragen
```

```python
serie
```

```python
data = pd.read_csv("Data\PopulationData.csv", delimiter=";", decimal=".", thousands=",", index_col=0, usecols=lambda x: x != "#")  #.drop(columns="#")
```

Die dropna() Funktion kann noch konfiguriert werden

```python
data.dropna(how='any')  # Lösche alle Felder bei denen es mindestens einen NaN Wert gibt (how='all' für NaN Werte)
```

```python
data.dropna(thresh=2)  # Lösche alle Zeilen die mindestens zwei Fehler haben
```

## Fehlerhafte Werte ersetzen

Es gibt mehrere Möglichkeiten fehlerhafte Werte zu ersetzen. Es kommt dabei auf das Datenset an, welche neuen Werte Sinn machen.

```python
data.fillna(0)  # Füllt jeden NaN Wert aus
```

```python
data["Migrants(net)"] = data["Migrants(net)"].fillna(0)  # Einzelne Spalte ausfüllen
```

```python
data.tail(50)
```

```python
data["Migrants(net)"].fillna(method="bfill")  # Übernimmt den Wert des Datensatzes mit dem oberen nicht-leeren Wert
data["Migrants(net)"].fillna(method="ffill")  # Übernimmt den Wert des Datensatzes mit dem unteren nicht-leeren Wert
```

## Fehlerhafte andere Werte ersetzen

Oftmals werden Fehler durch Formatierung verursacht (z.B.: Prozentzeichen an Ende). Auch diese können ausgebessert werden, allerdings gibt es dafür keine direkte Methode.

```python
data["Med.Age"] == "N.A."
```

```python
data[data["Med.Age"] == "N.A."]
```

```python
data.loc[data["Med.Age"] == "N.A.", "Med.Age"] = 0  # 0 ist hier nicht unbedingt der beste Wert
```

```python
data
```

```python
filtered = data[data["Med.Age"] != "N.A."]
mean = filtered["Med.Age"].astype(int).mean()
data.loc[data["Med.Age"] == "N.A.", "Med.Age"] = mean
```

```python
data
```

```python
data.info()
```

```python
data = data.convert_dtypes()  # Automatische Konvertierung (funktioniert nicht immer)
```

```python
data.info()
```

```python
data["Med.Age"] = data["Med.Age"].astype(float)
```

```python
data
```

```python
data.loc[data["Fert.Rate"] == "N.A.", "Fert.Rate"] = "0"
```

```python
data["Fert.Rate"] = data["Fert.Rate"].astype(float)
```

```python
data
```

```python
data["YearlyChange"].replace(" %", "", regex=True)  # regex=True um Teile der Werte zu ersetzen (sonst werden nur ganze Werte angeschaut)
```

```python
data["YearlyChange"] = data["YearlyChange"].replace(" %", "", regex=True)
```

```python
data
```

```python
def replacePct(dataframe, colName):
    dataframe[colName] = dataframe[colName].replace(" %", "", regex=True)
```

```python
replacePct(data, "UrbanPop %")
```

```python
data
```

```python
replacePct(data, "WorldShare")
```

```python
data
```

```python
data["UrbanPop %"] = data.loc[data["UrbanPop %"] == "N.A.", "UrbanPop %"] = 0
```

```python
data.convert_dtypes()
```

```python
data.info()
```

```python
data["UrbanPop %"] = data["UrbanPop %"].astype(float)
```

```python
data["YearlyChange"] = data["YearlyChange"].astype(float)
```

```python
data["WorldShare"] = data["WorldShare"].astype(float)
```
