## Andere Datenquellen

Es ist auch möglich andere Quellen als nur CSV zu benutzen (z.B.: SQL oder JSON auch in Kombination mit einer API)

```python
import pandas as pd
import requests as api
```

```python
json = api.get("https://open.er-api.com/v6/latest/USD")
```

```python
json.json()
```

```python
json.json()["rates"]
```

```python
readJson = pd.DataFrame.from_dict(json.json()["rates"], orient="index")
```

```python
readJson
```

```python
readJson = readJson.rename(columns={0: "Exchange Rate"})
```

```python
readJson
```

```python
data = pd.read_json("https://open.er-api.com/v6/latest/USD")
```

```python
data
```

## Daten aus einer Datenbank einlesen

```python
import sqlite3 as sql
connection = sql.connect("Test2.db")

cursor = connection.cursor()

cursor.execute("create table test(id integer primary key, name varchar(30))")
cursor.execute("insert into test (id, name) values(1, 'Max1')")
cursor.execute("insert into test (id, name) values(2, 'Max2')")
cursor.execute("insert into test (id, name) values(3, 'Max3')")

data = pd.read_sql("select * from test", con=connection)
```

```python
data
```
