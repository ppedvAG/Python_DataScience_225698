## Matplotlib Grundlagen

Es gibt zwei Modi für Graphen: der globale und der Objekteorientierte Modus

```python
# Globaler Modus
a = np.arange(100)
plt.figure(figsize=(12, 6))
plt.title("Zahlen von 0 bis 100, mal zwei und mal drei")
plt.plot(a)
plt.plot(a * 2)
plt.plot(a * 3)
```

Das ist der globale Modus, hier kann man einfach Werte setzen durch verschiedene Methoden und danach wird der Graph gezeichnet.

Hier können auch mehrere Werte in einem einzelnen Graph sein.

```python
x = np.arange(-10, 11)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)  # Anzahl Zeilen, Anzahl Spalten, Nummer die jetzt konfiguriert wird
plt.plot(x, x ** 2)
plt.plot([0, 0, 0], [-10, 0, 100])
plt.legend(["X^2", "Vertikal"], loc="lower right")
plt.xlabel("X")
plt.ylabel("X^2")

plt.subplot(1, 2, 2)  # Anzahl Zeilen, Anzahl Spalten, Nummer die jetzt konfiguriert wird
plt.plot(x, -x ** 2)
plt.legend(["-X^2"], loc="lower center")
plt.xlabel("X")
plt.ylabel("-X^2")
```

## Objektorientierter Modus

Der Objektorientierter Modus ermöglicht uns die Graphen sehr genau zu konfigurieren.

```python
fig, axes = plt.subplots(figsize=(12, 6))  # Subplots -> Tupel mit Figure und Axis
```

```python
f"{type(fig)}{type(axis)}"
```

```python
# Erforderlich: Die Daten selbst (x), Form der Daten (x ** 2)
axes.plot(
    x, x ** 2,
    color="red",
    linewidth=3,
    marker="o",
    markersize=8,
    label="X^2")

axes.plot(
    x, (x ** 2) * -1,
    linestyle='--',
    label="-X^2")

axes.set_xlabel("X")
axes.set_ylabel("Y")
axes.set_title("Ein Plot")

axes.legend()

fig
```

```python
fig, axes = plt.subplots(figsize=(12, 6))

axes.plot(x, x * 1, linestyle="solid")
axes.plot(x, x * 2, linestyle="dashed")
axes.plot(x, x * 3, linestyle="dashdot")
axes.plot(x, x * 4, linestyle="dotted")
```

## Arbeiten mit mehreren OOP-Plots

Bei der subplots() Funktion können auch die Dimensionen angegeben werden.

```python
fig, ((a1_1, a1_2), (a2_1, a2_2)) = plt.subplots(nrows=2, ncols=2, figsize=(9, 9))  # Hier wichtig: Tupel als Achsen statt 4 einzelne Werte
```

```python
a1_1.plot(np.random.randn(50), c="green")
a1_2.plot(np.random.randn(50), c="yellow")
a2_1.plot(np.random.randn(50), c="red")
a2_2.plot(np.random.randn(50), c="blue")

fig
```

## Bar Plot (Balkendiagramm)

Zeigt absolute Daten mit Bars an.

```python
x = np.array([np.random.randint(0, 10) for _ in range(0, 10)])
y = np.array([np.random.randint(0, 10) for _ in range(0, 10)])

plt.figure(figsize=(12, 6))
# Erforderlich: Ticks in Höhe der Anzahl der X Werte (Liste von Werten), Werte (x)
plt.bar(np.arange(0, len(x)), x, width=0.4, color="#00FF00", label="X")  # barh für horizontaler Chart
plt.bar(np.arange(0, len(y)) + 0.4, y, width=0.4, color="#0000FF", label="Y")
plt.grid(color="000000", alpha=0.2, axis="y", linestyle="--", linewidth=2, label="Grid")
plt.xlabel("Index")
plt.ylabel("Zahlenwerte")
#plt.xticks([i for i in range(0, 10)])
plt.legend()
```

## Histogramm

## Pie Chart (Kreisdiagramm)

Zeigt die Daten in einer Anteilsmenge an.

```python
x = np.arange(1, 10)

plt.figure(figsize=(6, 6))

# Erforderlich: x
plt.pie(x,
        #radius=1.5,
        #center=(4, 4),
        frame=True,
        labels=[f"Zahl: {i}" for i in x],
        labeldistance=0.6,
        counterclock=False,
        startangle=180,
        rotatelabels=True
       )

plt.xticks([])
plt.yticks([])

plt.xlabel("X")
plt.ylabel("Y")

None
```

## Scatterplot (Streudiagramm)

Ein Streudiagramm erlaubt es uns, zwei Werte gleichzeitig zu visualisieren.

Diesen Diagramm wird vorallem dazu verwendet um Outlier zu erkennen.

```python
x = [i * (1 + np.random.rand()) for i in range(0, 100)]  # Datenpunkte definieren
y = [i * (1 + np.random.rand()) for i in range(0, 100)]

col = [x[i] * y[i] for i in range(0, 100)]  # Farben angeben
area = [i * 2 for i in range(0, 100)]  # Fläche für einzelne Punkte angeben

plt.scatter(x, y, c=col, s=area)
plt.colorbar()

None
```
